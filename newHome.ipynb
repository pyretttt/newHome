```python
import requests
import pandas as pd

import sys
sys.path.append('../newHome/')
from Model import ModelEntity
```

Конфигурация

```python
MAX_LIST_LIMIT_TO_FETCH = 1000
MAX_OBJECT_TO_FETCH = 5
```

```python
def getObjectListURL(limit: int):
    return "https://наш.дом.рф/сервисы/api/kn/object?offset=0&limit=%d&sortField=devId.devShortCleanNm&sortType=asc&objStatus=0" % limit

def getObjectByID_URL(identifier: int):
    return f"https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/object/{identifier}"
```

Получаем общее кол-во объектов:

```python
response = requests.get(getObjectListURL(0))
print(response)
ITEMS_COUNT = response.json()["data"]["total"]
```

```python
list_limit = ITEMS_COUNT if MAX_LIST_LIMIT_TO_FETCH is None else MAX_LIST_LIMIT_TO_FETCH

objects_response = requests.get(getObjectListURL(list_limit)).json()
print(objects_response)
```

Формируем список идентификаторов объектов

```python
object_identifiers = list()
for object_container in objects_response.get("data").get("list"):
    object_identifiers.append(object_container.get("objId"))

print(object_identifiers)
```

```python
item_to_fetch_limit = len(object_identifiers) if MAX_OBJECT_TO_FETCH is None else MAX_OBJECT_TO_FETCH

items_responses = list()
for item in range(item_to_fetch_limit):
    item_response = requests.get(getObjectByID_URL(object_identifiers[item]))
    items_responses.append(item_response.json()["data"])
    print(item_response.json())
```

```python
models = [ModelEntity(item).data for item in items_responses]

data_frame = pd.DataFrame(models)
print(data_frame)
```

Сохраняем в файлы

```python
data_frame.to_excel("./output.xlsx")
data_frame.to_pickle("./output.pkl")
```
